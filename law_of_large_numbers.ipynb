```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Setup
# I'm defining the parameters of the experiment here.
# We are simulating flipping a fair coin.
N_TRIALS = 5000  # How many times I'll flip the coin in total
THEORETICAL_PROB = 0.5 # A fair coin should land on heads 50% of the time.

print(f"Starting simulation of {N_TRIALS} coin flips...")

# 2. The Simulation
# Instead of a slow Python loop, I'm using NumPy to generate all flips at once.
# We generate N_TRIALS random integers, either 0 (Tails) or 1 (Heads).
outcomes = np.random.randint(0, 2, N_TRIALS)

# 3. Calculating Cumulative Averages
# This is the crucial step for the visualization. I need to know the
# proportion of heads *at every step along the way*, not just at the end.

# First, count total heads accumulated up to each point.
cumulative_heads = np.cumsum(outcomes)

# Next, create an array representing the trial number (1st flip, 2nd flip... Nth flip).
trial_numbers = np.arange(1, N_TRIALS + 1)

# Calculate the running proportion: (total heads so far) / (total trials so far)
running_proportions = cumulative_heads / trial_numbers


# 4. Visualization
# Time to plot the results to see the law in action.
plt.figure(figsize=(12, 6))

# Plotting the actual experimental data (the jagged line)
plt.plot(trial_numbers, running_proportions, label='Experimental Proportion of Heads', color='#3498db', linewidth=1.5, alpha=0.8)

# Plotting the theoretical expected value (the straight line goalpost)
plt.axhline(y=THEORETICAL_PROB, color='#e74c3c', linestyle='--', linewidth=2, label=f'Theoretical Probability ({THEORETICAL_PROB})')

# Adding titles and labels to make the graph readable
plt.title(f'Demonstration of the Law of Large Numbers\n({N_TRIALS} Simulated Coin Flips)', fontsize=14)
plt.xlabel('Number of Trials (Coin Flips)', fontsize=12)
plt.ylabel('Proportion of Heads', fontsize=12)
plt.legend(loc='best', frameon=True)
plt.grid(True, which='both', linestyle='--', alpha=0.5)

# Setting y-axis limits to focus on the convergence area
plt.ylim(0.3, 0.7)

# Show the final result
print("Simulation complete. Generating plot...")
plt.show()
```
